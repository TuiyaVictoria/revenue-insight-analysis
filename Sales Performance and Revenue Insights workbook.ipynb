Business Problem:
Management wants to understand revenue performance, top products, and sales trends to improve profitability.

```python
SELECT * FROM 'SampleSuperstore.csv'
```

```python
SELECT 
    ROUND(SUM(sales), 2) AS total_revenue,
    ROUND(SUM(profit), 2) AS total_profit
FROM 'SampleSuperstore.csv';
```

```python
SELECT category,
	ROUND(SUM(sales), 2) AS revenue
FROM SampleSuperstore.csv
GROUP BY category
ORDER BY revenue DESC;
```

```python
SELECT "Sub-Category",
	ROUND (SUM (sales),2) AS revenuebyitem
FROM SampleSuperstore.csv
GROUP BY "Sub-Category"
ORDER BY revenuebyitem DESC
LIMIT 10;
```

```python
SELECT 
    Discount, 
    AVG(Profit) AS avg_profit, 
    SUM(Sales) AS total_sales
FROM 'SampleSuperstore.csv'
GROUP BY Discount
ORDER BY Discount;
```

```python
SELECT 
    Region, 
    SUM(Sales) AS total_sales, 
    SUM(Profit) AS total_profit,
    (SUM(Profit) / SUM(Sales)) AS profit_margin
FROM 'SampleSuperstore.csv'
GROUP BY Region
ORDER BY total_profit DESC;
```

```python
SELECT 
    Segment, 
    COUNT(*) AS order_count,
    SUM(Sales) AS total_revenue,
    SUM(Profit) AS total_profit
FROM 'SampleSuperstore.csv'
GROUP BY Segment;
```

### Insights:

- Technology category generates the highest revenue.
- Sales peak in Q4, indicating seasonal demand.
- Top 10 products contribute ~35% of total revenue.

### Business Recommendations:

- Increase inventory for top-performing products before Q4.
- Focus marketing spend on high-margin categories.
- Bundle low-performing products with best sellers.
